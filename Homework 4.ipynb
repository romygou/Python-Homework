Recall the simple random walk.  At each step, we flip a fair coin. If heads, we move "foward" one unit; if tails, we move "backward." 

## (A).

Way back in Homework 1, you wrote some code to simulate a random walk in Python. 

Start with this code, or write your own new code. Regardless, take your code, modify it, and enclose it in a function `rw()`. This function should accept a single argument `n`, the length of the walk. The output should be a list giving the position of the random walker, starting with the position after the first step. For example, 

```python
rw(5)
[1, 2, 3, 2, 3]
```

Unlike in the HW1 problem, you should not use upper or lower bounds. The walk should always run for as long as the user-specified number of steps `n`. 

Use your function to print out the positions of a random walk of length `n = 10`. 

Don't forget a helpful docstring! 

**Do not use ``numpy`` for this part**

````python
# solution (with demonstration) here

def rw(n):
    '''
    A function that simulates a simple random walk and outputs a list of the resulting positions
    
    Parameters
    ----------
    n: int, the length of the walk
    
    Return
    ----------
    positions: list, a list of the positions of the random walker, starting with the position after the first step
    '''
    
    import random

    #initializations
    steps = 0
    positions = []
    walk_len = 0

    while True:
        x = random.choice([1,-1]) #use 1 and -1 instead of heads and tails to make code more concise
        walk_len += 1 
        steps += x
        positions.append(steps) #append current position to list that logs positions
        if walk_len == n: #if walk length is reached, break loop and return list
            return(positions)

#demonstration
rw(10)
````

Now create a function called `rw2(n)`, where the argument `n` means the same thing that it did in Part A. Do so using `numpy` tools. Demonstrate your function as above, by creating a random walk of length 10. You can (and should) return your walk as a `numpy` array.

````python
# solution (with demonstration) here

import numpy as np

#create an array of size 10 containing a random distribution of -1s and 1s
#find the cumulative sum of the array
rw2 = lambda n: np.random.choice([-1, 1], size = n).cumsum()

#demonstration
rw2(10)
````
